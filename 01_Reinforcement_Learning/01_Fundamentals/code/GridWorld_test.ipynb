```python
import sys
sys.path.append('./src')

import GridWorld 
```

```python
# 1. 정상 케이스
print("=== 정상 케이스 ===")
env = GridWorld(size=5, obstacles=[(1,1), (2,2), (3,1)])
env.reset()
env.render()
```

```python

# 2. obstacles=None
print("=== obstacles=None ===")
env2 = GridWorld(size=5)
env2.reset()
env2.render()

# 3. assert 케이스
print("=== assert 케이스 ===")
try:
    env3 = GridWorld(size=5, obstacles=[(0,0)])
except AssertionError as e:
    print(f"AssertionError: {e}")

try:
    env4 = GridWorld(size=5, obstacles=[(4,4)])
except AssertionError as e:
    print(f"AssertionError: {e}")

# 4. step 테스트
print("=== step 테스트 ===")
env5 = GridWorld(size=5, obstacles=[(1,1)])
env5.reset()
for action in [1, 3, 3, 1, 3]:  # 몇 번 이동
    state, reward, done = env5.step(action)
    print(f"state: {state}, reward: {reward}, done: {done}")
env5.render()

# 5. 경계 테스트
print("=== 경계 테스트 ===")
env6 = GridWorld(size=5)
env6.reset()
for _ in range(5):  # 왼쪽 벽에 계속 부딪히기
    state, reward, done = env6.step(2)  # left
    print(f"state: {state}, reward: {reward}")
```
