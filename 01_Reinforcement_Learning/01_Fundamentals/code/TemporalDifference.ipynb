## 4-6. Comparing SARSA, Expected SARSA, Q-Learning, Double Q-Learning

```python
bigger_env = GridWorld(
    size=10,
    obstacles=[(1,1),(1,2),(2,4),(3,7),(4,2),(5,5),(6,1),(6,8),(7,3),(8,6)],
    icy_floors=[(2,3),(3,5),(4,7),(5,2),(6,6),(7,8),(8,3)]
)

td_compare = TD0(bigger_env)
```

```python
import matplotlib.pyplot as plt

_, Q_sarsa, hist_sarsa, step_sarsa = td_compare.SARSA(max_iter=10000)
_, Q_esarsa, hist_esarsa, step_esarsa = td_compare.ESARSA(max_iter=10000)
_, Q_ql, hist_ql, step_ql = td_compare.QL(max_iter=10000)
_, Q_dql, hist_dql, step_dql = td_compare.DQL(max_iter=10000)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(hist_sarsa, label="SARSA")
plt.plot(hist_esarsa, label="Expected SARSA")
plt.plot(hist_ql, label="Q-learning")
plt.plot(hist_dql, label="Double Q-learning")
plt.xlabel("Episode")
plt.ylabel("Mean Q")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(step_sarsa, label="SARSA")
plt.plot(step_esarsa, label="Expected SARSA")
plt.plot(step_ql, label="Q-learning")
plt.plot(step_dql, label="Double Q-learning")
plt.ylabel("Steps to goal")
plt.xlabel("Episode")
plt.xlim((0, 100))
plt.legend()
plt.grid()
plt.show()
```
