## 4-1. Evaluation

```python
target_policy = np.random.dirichlet(np.ones(len(env.actions)), size = len(env.states))

td0 = TD0(env)

Opt_V_TD0 = td0.evaluation(target_policy)

print(f"Optimal V using TD0:\n{Opt_V_TD0.reshape(5, 5)}")
```

## 4-2. SARSA

```python
Opt_p_SARSA, Opt_Q_SARSA = td0.SARSA(alpha = 0.1, gamma = 0.99, max_steps=1000, max_iter=100000)

print("========= Optimal Policy - SARSA =========")
print(Opt_p_SARSA)
print("\n========= Optimal Q - SARSA =========")
print(Opt_Q_SARSA)
```
